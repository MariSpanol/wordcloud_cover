## Passo 2
Usando o texto que foi extraído para criar nossa wordcloud

```python
# importar os pacotes necessários
import numpy as np #formato da imagem
import pandas as pd #ler o txt
import matplotlib.pyplot as plt #gerar a WC como "foto"
from PIL import Image #Formatação da mascara
from wordcloud import WordCloud, STOPWORDS #Pacote para gerar a wc
 

```

```python
# Abrir o text e ver o conteudo
with open('about.txt', encoding='utf8') as f:
    for line in f:
        print(line)
```

```python
# lista de stopwords (aqueles vicios verbais que temos de monte no PT-BR)
stopwords = set(STOPWORDS)
stopwords.update(["da","e", "meu", "em", "você", "de", "ao", "os"])
```

```python
# gerar nossa wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=1600, height=800).generate(line)
```

#### Resultado da Wordcloud

```python
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

# salvar ela
wordcloud.to_file("cover_linkedin.png")
```

### Usando "mascara" na wordcloud
A mascara é aquele desenho bonitinho que fica atrás da wordcloud e aqui eu vou ensinar de um jeito facin transformar uma imagem qualquer (png) na certa para a nuvem de palavras

**Eu indico que o desenho seja branco e preto (em preto onde irão ficar as palavras, ou seja, background branco e desenho preto se você quiser que as palavras fiquem dentro do desenho e não no entorno)**

```python
# fazendo a conversao de png para jpg 
from PIL import Image

im = Image.open("lup.png")
rgb_im = im.convert('RGB')
rgb_im.save('myimg2.jpg')
rgb_im
```

```python
# lendo a imagem transformada e olhando o array transformado em RGB
mask = np.array(Image.open("myimg2.jpg"))
mask


   
```

```python
# Criando a Wordcloud 
#o argumento mask é onde eu passo a imagem
#O contour_color é a cor do contorno do desenho e contour_width é a "grossura" do contorno do desenho

wc = WordCloud(background_color="white", max_words=1000, mask=mask,
               stopwords=stopwords, contour_width=9, contour_color='pink').generate(line)
```

```python
# Vendo a imagem e salvando
plt.figure(figsize=[10,20])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

wc.to_file("wc.png")
```
