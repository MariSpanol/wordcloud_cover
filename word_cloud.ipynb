```python
# importar os pacotes necessários
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
 
# importar o arquivo csv em um df
df = pd.read_csv('about.txt')
 
# eliminar as colunas com valores ausentes
#summary = df.dropna(subset=['summary'], axis=0)['summary']
```

```python

with open('about.txt', encoding='utf8') as f:
    for line in f:
        print(line)
```

```python
# lista de stopword
stopwords = set(STOPWORDS)
stopwords.update(["da","e", "meu", "em", "você", "de", "ao", "os"])
```

```python
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(line)
 
```

```python
# gerar uma wordcloud
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="white",
                      width=1600, height=800).generate(line)
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

wordcloud.to_file("cover_linkedin.png")
```
